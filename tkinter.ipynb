```python
from tkinter import *

main=Tk()
main.title("Calculator")
main.geometry("200x200")

Label(main,text="Enter number").grid(row=0,column=0,sticky=W)
Label(main,text="Enter second number").grid(row=1,column=0,sticky=W)
Label(main,text="Output").grid(row=2,column=0,sticky=W)

e1=Entry(main,width=20,bg="light grey")
e1.grid(row=0, column=1,sticky=W)

e2=Entry(main,width=20,bg="light grey")
e2.grid(row=1, column=1,sticky=W)

def plus():
    n1=int(e1.get())
    n2=int(e2.get())
    res=n1+n2
    o1.insert(END,res)

def minus():
    n1=int(e1.get())
    n2=int(e2.get())
    res=n1-n2
    o1.insert(END,res)

def mul():
    n1=int(e1.get())
    n2=int(e2.get())
    res=n1*n2
    o1.insert(END,res)

def div():
    n1=int(e1.get())
    n2=int(e2.get())
    res=n1/n2
    o1.insert(END,res)

def clear():
    e1.delete(0,END)
    e2.delete(0,END)
    o1.delete(1.0,END)

o1=Text(main,width=20,height=2)
o1.grid(row=2,column=1,sticky=W)

Button(main,text="+",command=plus,width=10).grid(row=3,column=0,sticky=W)
Button(main,text="-",command=minus,width=10).grid(row=3,column=1,sticky=W)
Button(main,text="*",command=mul,width=10).grid(row=4,column=0,sticky=W)
Button(main,text="/",command=div,width=10).grid(row=4,column=1,sticky=W)
Button(main,text="Clear",command=clear,width=10).grid(row=5,column=1,sticky=W)

main.mainloop()
```

```python
class A:
    def m(self):
        print("m of A called")

class B(A):
    def m(self):
        print("m of B called")
        super().m()
    
class C(A):
    def m(self):
        print("m of C called")
        super().m()

class D(B,C):
    def m(self):
        print("m of D called")
        super().m()

x=D()
x.m()
```
